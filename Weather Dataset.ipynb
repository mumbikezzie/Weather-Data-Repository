```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv(r'C:\\Users\LENEVO X240\Downloads\archiveweather\1. Weather Data.csv')
```

```python
dataset.head()
```

```python
#1. Find the records where the weather is exactly clear

clear_weather=dataset[dataset['Weather'] =='Clear']

#Display results
print(clear_weather)
```

```python
#2. Find the number of times the wind speed was exactly 4 km/hr.

#filter the dataset for wind speed exactly 4km/hr
wind_speed_4=dataset[dataset['Wind Speed_km/h']==4]

#count the number of occurences
count_wind_speed_4=len(wind_speed_4)

#display the result
print(f"The wind speed was exactly 4km/hr {count_wind_speed_4} times.")

#display records where the wind speeds were 4km/h
print("\nRecords with wind speeds 4km/hr:")
print(wind_speed_4)
```

```python
#3. Check if there are any null values present in the dataset


#checking for null values
null_counts = dataset.isnull().sum()

#displaying the count of null values for each column
print("Count of null values in each column:")
print(null_counts)

# Display total number of null values
total_null = dataset.isnull().sum().sum()
print(f"\nTotal number of null values in the dataset: {total_null}")

# Display if there are any null values
if total_null > 0:
    print("There are null values present in the dataset.")
else:
    print("There are no null values in the dataset.")
```

```python
#4. Rename the column "Weather" to "Weather_Condition"

dataset = dataset.rename(columns={'Weather': 'Weather_Condition'})

# Verify the change
print(dataset.columns)

# Display the first few rows to confirm the change
print(dataset[['Weather_Condition']].head())
```

```python
#5. What is the mean visibility of the dataset?

# Find the mean visibility
mean_visibility = dataset['Visibility_km'].mean()

# Display the result
print(f"The mean visibility is {mean_visibility:.2f} km")

# Calculate mean, ignoring NaN values
mean_visibility_handled = dataset['Visibility_km'].mean()

print(f"The mean visibility (handling potential non-numeric values) is {mean_visibility_handled:.2f} km")
```

```python
#6. Find the number of records where wind speed is greater than 24km/hr and visibility is equal to 25km

#filter the dataset
filtered_data = dataset[(dataset['Wind Speed_km/h'] > 24 ) & (dataset['Visibility_km'] == 25)]

#count the number of records
count=len(filtered_data)

#Display the results
print(f"Number of records where wind speed is greater than 24km/hr & visibility = 25km: {count}" )
```

```python
#7. What is the mean value of each column for each weather condition?

#making sure all columns except weather condition are numeric
numeric_columns = dataset.columns.drop('Weather_Condition')
dataset[numeric_columns] = dataset[numeric_columns].apply(pd.to_numeric, errors='coerce')

# group columns by weather condition and calculate mean for each numeric column:
mean_by_weather_condition = dataset.groupby('Weather_Condition')[numeric_columns].mean()

#Display the results
print("Mean values for each column by weather condition:")
print(mean_by_weather_condition)

#total mean
total_mean= dataset[numeric_columns].mean()
print("\nTotal mean values:")
print(total_mean)
```

```python
#8. Find all instances where the weather is clear and the relative humidity is greater than 50, OR visibility is above 40

#filter the dataset
filtered_data=dataset[
    ((dataset['Weather_Condition']=='Clear') & (dataset['Rel Hum_%']>50)) |
    ((dataset['Visibility_km']>40))
]

#count the number of instances
count=len(filtered_data)

#Print the result
print(f"Number of records that have over 50 relative humidity or visibility above 40: {count}")
```

```python
#9. Find the number of weather conditions that include snow

#Checking to see if 'snow' is part of the weather conditions:
def contains_snow(weather_condition):
    return 'snow' in weather_condition.lower()

#Apply the function to the 'weather_condition' column
snow_mask = dataset['Weather_Condition'].apply(contains_snow)

#Count number of records with snow
snow_count = snow_mask.sum()

#Get unique weather conditions with snow
snow_conditions =dataset.loc[snow_mask, 'Weather_Condition'].unique()

#Print the results
print(f"Number of records with snow: {snow_count}")
print("\nUnique weather conditions including snow:")
for condition in snow_conditions:
    print (condition)
    
    




```
